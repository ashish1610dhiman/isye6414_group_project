## Getting artist popularity and followers

```python
data = pd.read_csv('../top_500_artists_deduplicate_genres.csv')
```

```python
data.head(5)
```

```python
unique_artists = pd.unique(data.artist).tolist()
```

```python
artist_df = []
for i in unique_artists:
    response = sp.search(i, type="artist", limit=1)
    artist_id = get_artist_id(response)
    artist_df.append({'name': i, 'artist_id': artist_id})
```

```python
artists = pd.DataFrame(artist_df)
```

```python
artist_ids = artists.artist_id.tolist()
```

```python
len(artist_ids)
```

```python
np_artists_id_batches = np.array_split(np.array(artist_ids), 20)
```

```python
meta_data = []
for i in np_artists_id_batches:
    artists_meta = sp.artists(i)
    for i in artists_meta['artists']:
        meta_data.append({'id': i['id'], 'popularity': i['popularity'], 'followers': i['followers']['total']})
```

```python
meta_data
```

```python
meta_df = pd.DataFrame(meta_data)
```

```python
meta_df.shape
```

```python
meta_df = meta_df.drop_duplicates(subset=['id'])
```

```python
meta_df.shape
```

```python
meta_df.head(5)
```

```python
meta_df[meta_df.id == '1ZwdS5xdxEREPySFridCfh']
```

```python
artist_meta = artists.merge(meta_df, left_on='artist_id', right_on='id', how='right')
```

```python
artist_meta = artist_meta.drop(['id'], axis=1)
```
